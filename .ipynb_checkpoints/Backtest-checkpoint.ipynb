```python
def model(T, beta_0, beta_1, beta_2, lambd):
    term1 = (1 - np.exp(-lambd * T)) / (lambd * T)
    term2 = term1 - np.exp(-lambd * T)
    Y = beta_0 + beta_1 * term1 + beta_2 * term2
    return Y
```

```python
backtest['beta_0_pred'] = backtest['beta_0'].rolling(window=3).mean().shift(1)
backtest['beta_1_pred'] = backtest['beta_1'].rolling(window=3).mean().shift(1)
backtest['beta_2_pred'] = backtest['beta_2'].rolling(window=3).mean().shift(1)

```

```python
columns_to_predict = ['DGS3MO', 'DGS6MO', 'DGS1', 'DGS2', 'DGS3', 'DGS5', 'DGS7', 'DGS10']

# Loop through each column name and create a new column with '_pred' suffix, initialized with 0
for column in columns_to_predict:
    backtest[f'{column}_pred'] = 0
    backtest[f'{column}_pos'] =0
    backtest[f'{column}_strategy_return'] =0
    backtest[f'{column}_cum_strat'] =0
    backtest[f'{column}_log_return'] = np.log(backtest[f'{column}'] / backtest[f'{column}'].shift(1))

```

```python
backtest
```

```python
for row in range(3,len(backtest)):
    y_data = backtest.iloc[row].values[1:]
    
    x_data = np.array([0.25,0.5,1, 2, 3, 5, 7, 10])
    y_pred = model(x_data, y_data[12] , y_data[13], y_data[14] ,y_data[11])
    #print(y_pred)
    i = 0
    for column in columns_to_predict:
        backtest.loc[row,f'{column}_pred'] = y_pred[i]
        i +=1
        if backtest.loc[row,f'{column}_pred'] < backtest.loc[row,f'{column}']:
            backtest.loc[row,f'{column}_pos'] = -1
        else:
            backtest.loc[row,f'{column}_pos'] = 1

        backtest.loc[row,f'{column}_strategy_return'] = backtest[f'{column}_log_return'][row] * backtest[f'{column}_pos'][row]

for column in columns_to_predict:
    backtest[ f'{column}_cum_strat'] = backtest[f'{column}_strategy_return'].cumsum().apply(np.exp)

        
            
    
    
    
```
